## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
   For Example "Landing_Outcome"**

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
task1 = cur.execute("SELECT DISTINCT Launch_Site FROM SPACEXTBL").fetchall()
print(task1)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
task2 = cur.execute("SELECT * FROM SPACEXTBL WHERE Launch_Site LIKE 'CCA%' LIMIT 5").fetchall()
task2 = pd.DataFrame(task2,columns=column)
task2
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
task3=cur.execute("SELECT SUM(PAYLOAD_MASS__KG_) FROM SPACEXTBL WHERE Customer ='NASA (CRS)'").fetchall()
task3
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
task4 = cur.execute("SELECT AVG(PAYLOAD_MASS__KG_) FROM SPACEXTBL WHERE Booster_Version = 'F9 v1.1'").fetchall()
task4
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
task5 = cur.execute("SELECT [Landing _Outcome],MIN(Date) FROM SPACEXTBL WHERE [Landing _Outcome] LIKE 'Success (ground pad)'").fetchall()
task5
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
landing = cur.execute("SELECT DISTINCT([Landing _Outcome]) FROM SPACEXTBL").fetchall()
landing
```

```python
task6 = cur.execute("SELECT DISTINCT(Booster_Version) FROM SPACEXTBL WHERE PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000 AND [Landing _Outcome] = 'Success (drone ship)'").fetchall()
task6
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
task7_1 =cur.execute("SELECT COUNT(Mission_Outcome) FROM SPACEXTBL WHERE Mission_Outcome LIKE '%Success%'").fetchall()
task7_1
```

```python
task7_2 =cur.execute("SELECT COUNT(Mission_Outcome) FROM SPACEXTBL WHERE Mission_Outcome LIKE '%Failure%'").fetchall()
task7_2
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
task8 = cur.execute("SELECT Booster_Version, PAYLOAD_MASS__KG_ FROM SPACEXTBL WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL)").fetchall()
task8
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
task9=cur.execute("SELECT Date,substr(Date,4,2),[Landing _Outcome],Booster_Version,Launch_Site FROM SPACEXTBL WHERE [Landing _Outcome]='Failure (drone ship)' AND substr(Date,7,4)='2015'").fetchall()
task9
```

### Task 10




##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python
task10=cur.execute("SELECT [Landing _Outcome],COUNT(*) FROM SPACEXTBL WHERE DATE BETWEEN '04-06-2010' AND '20-03-2017' GROUP BY [Landing _Outcome] ORDER BY COUNT([Landing _Outcome]) DESC").fetchall()
task10
```
